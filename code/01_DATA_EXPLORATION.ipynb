```python
%matplotlib inline
import pandas as pd
import seaborn as sns
```

```python
train = pd.read_csv("../data/train.csv")
train.head()
```

```python
print("No of rows: {} - No of columns: {}".format(*train.shape))
print("Total is_trained == 0 : {} rows".format(train[train.is_listened == 0].shape[0]))
print("Total is_trained == 1 : {} rows".format(train[train.is_listened == 1].shape[0]))
```

# Basic Visualisation

## Content Type Distribution

```python
%time
plot = sns.countplot(x="context_type", hue="is_listened", data=train)
plot.figure.set_size_inches(20,10)
```

## Platform Name Distribution

```python
%time
plot = sns.countplot(x="platform_name", hue="is_listened", data=train)
plot.figure.set_size_inches(20,10)
```

## Platform Family Distribution

```python
%time
plot = sns.countplot(x="platform_family", hue="is_listened", data=train)
plot.figure.set_size_inches(20,10)
```

## Listen Type Distribution

```python
%time
plot = sns.barplot(x="listen_type", y="is_listened", data=train)
plot.figure.set_size_inches(20,10)
```

## User Gender Distribution

```python
%time
plot = sns.countplot(x="user_gender", hue="is_listened", data=train)
plot.figure.set_size_inches(20,10)
```

## User Age Distribution

```python
%time
plot = sns.countplot(x="user_age", hue="is_listened", data=train)
plot.figure.set_size_inches(20,10)
```
